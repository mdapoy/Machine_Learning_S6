## Data preparation

### Subtask:
Prepare the data for clustering by handling outliers and scaling the features.


**Reasoning**:
Handle outliers using IQR method and then scale the features using StandardScaler.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Calculate IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_filtered = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

# Scale the features using StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_filtered), columns=df_filtered.columns)

display(df_scaled.head())
display(df_scaled.describe())
```

## Data clustering

### Subtask:
Implement two clustering algorithms, K-Means and DBSCAN, on the preprocessed data.


**Reasoning**:
Implement K-Means and DBSCAN clustering algorithms on the preprocessed data (df_scaled).



```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

# K-Means Clustering
inertia_values = []
silhouette_scores = []
kmeans_labels = []
k_range = range(2, 11)  # Example k range

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_scaled, kmeans.labels_))
    kmeans_labels.append(kmeans.labels_)

# DBSCAN Clustering
dbscan_labels = []
eps_values = np.arange(0.1, 2.1, 0.1)  # Example eps range
min_samples = 5

for eps in eps_values:
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan.fit(df_scaled)
    dbscan_labels.append(dbscan.labels_)

print("KMeans inertia values:", inertia_values)
print("KMeans silhouette scores:", silhouette_scores)
print("DBSCAN labels:", dbscan_labels)
```

## Model evaluation

### Subtask:
Evaluate the K-Means and DBSCAN clustering results using the Elbow method, Silhouette Coefficient, Calinski-Harabasz Index, Davies-Bouldin Index, and Dunn Index.  Explain the mathematical equations behind each metric in Indonesian. Determine the "best" metric and justify your choice.


**Reasoning**:
Evaluate the K-Means and DBSCAN clustering results using the specified metrics, provide explanations of the mathematical equations, and determine the best metric.



```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances

# Function to calculate Dunn index
def dunn_index(points, labels):
    """Calculates the Dunn index for a given clustering result."""
    n_clusters = len(set(labels))
    if n_clusters == 1:
        return 0  # Avoid division by zero when only one cluster

    distances = pairwise_distances(points)
    cluster_indices = [[] for _ in range(n_clusters)]
    for i, label in enumerate(labels):
        cluster_indices[label].append(i)

    min_intercluster_distance = float('inf')
    for i in range(n_clusters):
        for j in range(i + 1, n_clusters):
            min_dist_clusters = float('inf')
            for point_i in cluster_indices[i]:
                for point_j in cluster_indices[j]:
                    min_dist_clusters = min(min_dist_clusters, distances[point_i, point_j])
            min_intercluster_distance = min(min_intercluster_distance, min_dist_clusters)

    max_intracluster_distance = 0
    for i in range(n_clusters):
        max_dist_cluster = 0
        for j in range(len(cluster_indices[i])):
            for k in range(j + 1, len(cluster_indices[i])):
                max_dist_cluster = max(max_dist_cluster, distances[cluster_indices[i][j], cluster_indices[i][k]])
        max_intracluster_distance = max(max_intracluster_distance, max_dist_cluster)

    if max_intracluster_distance == 0: # Handle the case where max_intracluster_distance is 0
      return float('inf')

    return min_intercluster_distance / max_intracluster_distance

# Evaluate K-Means
k_range = range(2, 11)
ch_scores = []
db_scores = []
dunn_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)

    ch_scores.append(calinski_harabasz_score(df_scaled, kmeans.labels_))
    db_scores.append(davies_bouldin_score(df_scaled, kmeans.labels_))
    dunn_scores.append(dunn_index(df_scaled.values, kmeans.labels_))

# Plotting and explanations will be added in the next step.

#Evaluate DBSCAN (Silhouette only, as other metrics are not directly applicable)
dbscan_silhouette_scores = []
for eps in np.arange(0.1, 2.1, 0.1):
    dbscan = DBSCAN(eps=eps, min_samples=5)
    dbscan.fit(df_scaled)
    labels = dbscan.labels_
    if len(set(labels)) > 1:  # Silhouette score requires more than one cluster
        dbscan_silhouette_scores.append(silhouette_score(df_scaled, labels))
    else:
        dbscan_silhouette_scores.append(0)

print("KMeans CH scores:", ch_scores)
print("KMeans DB scores:", db_scores)
print("KMeans Dunn scores:", dunn_scores)
print("DBSCAN silhouette scores:", dbscan_silhouette_scores)
```

```python
import matplotlib.pyplot as plt

# Plotting the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia_values, marker='o')
plt.title('Metode Elbow untuk K-Means')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

#Penjelasan Metode Elbow
print("Penjelasan Metode Elbow:\nMetode Elbow digunakan untuk menentukan jumlah cluster (k) yang optimal dalam algoritma K-Means.  Metode ini memplot nilai inertia (jumlah kuadrat jarak setiap titik data ke centroid cluster terdekatnya) terhadap berbagai nilai k. Titik siku pada plot menunjukkan nilai k yang meminimalkan inertia tanpa terlalu banyak menambah jumlah cluster, sehingga menjadi trade-off yang baik antara kompleksitas model dan kinerja cluster.")

#Penjelasan Rumus Inertia
print("\nPenjelasan Rumus Inertia:\nInertia dihitung sebagai jumlah kuadrat jarak Euclidean antara setiap titik data dan centroid cluster terdekat.  Rumusnya dapat dinyatakan sebagai berikut:")
print("Inertia = Î£áµ¢ Î£â±¼ ||xâ±¼ - cáµ¢||Â²")
print("di mana:")
print("xâ±¼ adalah titik data ke-j")
print("cáµ¢ adalah centroid cluster ke-i")
print("||.|| menunjukkan jarak Euclidean")


# Plotting Silhouette Scores for K-Means
plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score untuk K-Means')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Penjelasan Silhouette Coefficient
print("\nPenjelasan Silhouette Coefficient:\nSilhouette Coefficient mengukur seberapa mirip suatu titik data dengan clusternya sendiri dibandingkan dengan cluster lainnya.  Nilai Silhouette Coefficient berkisar antara -1 hingga 1. Nilai mendekati 1 menunjukkan bahwa titik data tersebut ditempatkan dengan baik di clusternya, nilai mendekati -1 menunjukkan bahwa titik data tersebut mungkin ditempatkan di cluster yang salah, dan nilai mendekati 0 menunjukkan bahwa titik data tersebut berada di antara dua cluster.")

# Penjelasan rumus Silhouette Coefficient
print("\nPenjelasan Rumus Silhouette Coefficient:")
print("Untuk setiap titik data i, Silhouette Coefficient dihitung sebagai:")
print("s(i) = (b(i) - a(i)) / max(a(i), b(i))")
print("di mana:")
print("a(i) adalah jarak rata-rata titik data i ke semua titik lain dalam cluster yang sama.")
print("b(i) adalah jarak rata-rata titik data i ke semua titik dalam cluster terdekat berikutnya.")

# Plot other metrics for KMeans
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot(k_range, ch_scores, marker='o')
plt.title('Calinski-Harabasz Index')
plt.xlabel('k')
plt.ylabel('CH Index')
plt.grid(True)

plt.subplot(1,3,2)
plt.plot(k_range, db_scores, marker='o')
plt.title('Davies-Bouldin Index')
plt.xlabel('k')
plt.ylabel('DB Index')
plt.grid(True)

plt.subplot(1,3,3)
plt.plot(k_range, dunn_scores, marker='o')
plt.title('Dunn Index')
plt.xlabel('k')
plt.ylabel('Dunn Index')
plt.grid(True)

plt.tight_layout()
plt.show()

#Penjelasan Calinski-Harabasz Index
print("\nPenjelasan Calinski-Harabasz Index:\nCalinski-Harabasz Index (atau Variance Ratio Criterion) mengukur seberapa baik cluster dipisahkan. Indeks ini lebih besar ketika cluster lebih padat dan terpisah dengan baik.")
#Penjelasan Rumus Calinski-Harabasz Index
print("\nPenjelasan Rumus Calinski-Harabasz Index:\n CH = (B/W) * ((n-k)/(k-1))")
print("di mana:\nB adalah jumlah variasi antar-cluster\nW adalah jumlah variasi dalam-cluster\nk adalah jumlah cluster\nn adalah jumlah sampel")

#Penjelasan Davies-Bouldin Index
print("\nPenjelasan Davies-Bouldin Index:\nDavies-Bouldin Index mengukur rata-rata kesamaan antara setiap cluster dan cluster terdekatnya. Nilai indeks yang lebih rendah menunjukkan clustering yang lebih baik.")
#Penjelasan Rumus Davies-Bouldin Index
print("\nPenjelasan Rumus Davies-Bouldin Index:")
print("Rumus Davies-Bouldin Index agak kompleks dan melibatkan perhitungan jarak antara centroid dan dispersi cluster.  Pada intinya, indeks ini mengukur seberapa 'dekat' cluster satu sama lain dan seberapa 'terpencar' data di dalam masing-masing cluster.  Nilai yang lebih rendah menunjukkan clustering yang lebih baik.")


#Penjelasan Dunn Index
print("\nPenjelasan Dunn Index:\nDunn Index mengukur kualitas clustering berdasarkan rasio jarak terkecil antar cluster dengan jarak terbesar di dalam cluster. Nilai yang lebih tinggi menunjukkan clustering yang lebih baik.")
#Penjelasan Rumus Dunn Index
print("\nPenjelasan Rumus Dunn Index:")
print("Dunn Index = min(jarak antar cluster) / max(jarak dalam cluster).")

# Plotting Silhouette Scores for DBSCAN
plt.figure(figsize=(8, 6))
plt.plot(np.arange(0.1, 2.1, 0.1), dbscan_silhouette_scores, marker='o')
plt.title('Silhouette Score untuk DBSCAN')
plt.xlabel('Epsilon (eps)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Discussion and justification of the best metric (in Indonesian)
print("\nPembahasan dan Pembenaran Metrik Terbaik:")
print("Berdasarkan hasil dan penjelasan di atas, kita dapat mengevaluasi metode clustering dengan memperhatikan karakteristik data dan tujuan analisis.  Pemilihan metrik terbaik sangat bergantung pada jenis data dan target clustering.")
print("Untuk dataset ini, perhatikan plot dan nilai-nilai yang dihasilkan dari setiap metrik.  Jika kita mencari cluster yang padat dan terpisah dengan jelas, maka Calinski-Harabasz Index bisa menjadi pilihan yang baik. Jika kita lebih memperhatikan pemisahan cluster, Dunn Index bisa menjadi pilihan lain.  Jika bentuk dan ukuran cluster tidak terlalu diperhatikan,  Silhouette Coefficient bisa menjadi pilihan yang baik.  Nilai Davies-Bouldin Index yang rendah juga mengindikasikan clustering yang baik.")

print("Karena DBSCAN cenderung menghasilkan cluster yang tidak teratur, Silhouette Coefficient tetap relevan untuk mengevaluasi kualitas cluster yang dihasilkan.")

# ... (rest of the analysis and conclusion) ...
```
