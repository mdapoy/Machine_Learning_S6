```python
import pandas as pd

# Fungsi untuk menghapus baris duplikat
def remove_duplicates(df, subset=None):
    """
    Menghapus baris yang duplikat dalam dataset.

    Parameters:
    df (pd.DataFrame): Dataset yang akan diproses
    subset (list, optional): Kolom-kolom yang digunakan untuk mengecek duplikat. Default adalah None yang berarti seluruh kolom.

    Returns:
    pd.DataFrame: Dataset tanpa baris duplikat
    """

    # Menghapus baris duplikat
    df_clean = df.drop_duplicates(subset=subset, keep='first')  # 'first' berarti menyimpan baris pertama, menghapus yang berikutnya

    return df_clean


# Menghapus baris duplikat berdasarkan seluruh kolom
df = remove_duplicates(df)
```

```python
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

# Misalkan dataset Anda ada di df, dengan fitur X dan label y
# X adalah fitur, y adalah label
X = df.drop('Class', axis=1)  # GanClass' dengan nama kolom label Anda
y = df['Class']  # Ganti 'labeClassngan nama kolom label Anda

# Tampilkan distribusi kelas sebelum undersampling
print("Distribusi kelas sebelum undersampling:", Counter(y))

# Inisialisasi RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy='auto', random_state=42)

# Terapkan undersampling
X_resampled, y_resampled = rus.fit_resample(X, y)

# Tampilkan distribusi kelas setelah undersampling
print("Distribusi kelas setelah undersampling:", Counter(y_resampled))
```

## MODELING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc

# Memisahkan data menjadi data pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# List model-model yang akan digunakan
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'KNeighbors': KNeighborsClassifier(),
    'Bagging': BaggingClassifier(),
    'Boosting': AdaBoostClassifier(),
    'SVM': SVC(probability=True)  # probability=True untuk menghitung AUC
}

# Menyimpan hasil evaluasi untuk setiap model
results = {}

# Melatih dan mengevaluasi setiap model
for name, model in models.items():
    # Melatih model
    model.fit(X_train, y_train)

    # Memprediksi nilai pada data uji
    y_pred = model.predict(X_test)

    # Menghitung metrik evaluasi
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Menyimpan hasil evaluasi
    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    }

    # Visualisasi confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Visualisasi AUC dan ROC
    if hasattr(model, "predict_proba"):  # Hanya model dengan prediksi probabilitas
        y_prob = model.predict_proba(X_test)[:, 1]  # Probabilitas untuk kelas positif
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)

        plt.figure()
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Receiver Operating Characteristic: {name}')
        plt.legend(loc="lower right")
        plt.show()

# Menampilkan hasil evaluasi
results_df = pd.DataFrame(results).T
print("Evaluasi Model:")
print(results_df)

# Menentukan model terbaik berdasarkan F1 score
best_model = results_df['F1 Score'].idxmax()
print(f'\nModel terbaik berdasarkan F1 Score: {best_model}')
```

## PENJELASAN

Penjelasan Kode:
Memisahkan Data: Menggunakan train_test_split untuk memisahkan dataset menjadi data pelatihan (X_train, y_train) dan data pengujian (X_test, y_test), dengan proporsi 70% pelatihan dan 30% pengujian.

Model yang Digunakan:

Logistic Regression: Model dasar untuk klasifikasi yang didasarkan pada regresi logistik.

Decision Tree: Model pohon keputusan yang membagi data berdasarkan fitur untuk membuat keputusan.

KNeighbors: Algoritma K-Nearest Neighbors (KNN), yang mengklasifikasikan berdasarkan kedekatannya dengan titik data lain.

Bagging: Teknik ensemble yang menggunakan model dasar untuk mengurangi varian model, seperti Random Forest.

Boosting: Model ensemble lain yang menggabungkan beberapa model yang lebih lemah untuk membentuk model yang lebih kuat, seperti AdaBoost.

SVM: Support Vector Machine, yang mencari hyperplane terbaik untuk memisahkan data antara kelas yang berbeda.

Evaluasi Model:

Setiap model dievaluasi dengan menggunakan beberapa metrik: Akurasi, Presisi, Recall, dan F1 Score.

Confusion Matrix: Menampilkan confusion matrix untuk setiap model untuk memvisualisasikan jumlah prediksi yang benar dan salah.

AUC dan ROC: Jika model mendukung probabilitas, kita juga akan menggambar kurva ROC dan menghitung AUC (Area Under the Curve).

Menentukan Model Terbaik: Berdasarkan nilai F1 Score, kita menentukan model terbaik. F1 Score adalah metrik yang sering digunakan untuk menilai performa model dalam klasifikasi dengan memerhatikan keseimbangan antara presisi dan recall.

Hasil yang Akan Ditampilkan:
Confusion Matrix: Menunjukkan jumlah prediksi yang benar dan salah untuk masing-masing kelas.

ROC Curve dan AUC: Menunjukkan performa model berdasarkan True Positive Rate (TPR) dan False Positive Rate (FPR).

Evaluasi Model: Menampilkan tabel dengan metrik evaluasi untuk setiap model.

Model Terbaik: Berdasarkan F1 Score, kita akan menampilkan model terbaik.

Analisis Model Terbaik:
Logistic Regression: Baik untuk dataset yang linear dan jika hubungan antar variabel tidak kompleks.

Decision Tree: Mudah diinterpretasi tetapi rentan terhadap overfitting jika tidak diatur dengan benar.

KNeighbors: Model sederhana, tetapi dapat sangat sensitif terhadap parameter dan ukuran data.

Bagging dan Boosting: Biasanya memberikan kinerja yang lebih baik karena mereka mengurangi overfitting (Bagging) atau meningkatkan bias (Boosting) dengan menggabungkan beberapa model.

SVM: Sering digunakan untuk data yang tidak linier dan berfungsi dengan baik pada data dimensi tinggi.

Model terbaik akan tergantung pada kinerja metrik di atas. Jika F1 Score adalah metrik utama, kita dapat memilih model dengan skor tertinggi di tabel tersebut.

Penjelasan untuk Model Terbaik:
F1 Score: Merupakan nilai yang menggabungkan Presisi dan Recall, memberikan gambaran yang lebih jelas tentang keseimbangan antara kedua metrik ini, terutama ketika kita memiliki dataset yang tidak seimbang (misalnya, kelas positif dan negatif tidak terdistribusi merata).

Jika satu model memiliki F1 Score lebih tinggi daripada yang lain, itu menunjukkan model tersebut lebih baik dalam menemukan keseimbangan antara presisi dan recall.
